```python
f = np.cos(x)
plt.plot(t, f, color="green")
plt.title("График f(t)", fontsize=20, fontweight="bold", color="#808080")
label_font = {
    "fontsize": 16,
    "family": "serif",
}
plt.xlabel("Значения t'",fontdict=label_font)
plt.ylabel("Значения f'",fontdict=label_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### Task 3

```python
x = np.linspace(-3,3,51)
print(x)
```

```python
y1 = x**2
print(y1)
```

```python
y2 = 2*x+0.5
print(y2)
```

```python
y3 = -3 * x - 1.5
print(y3)
```

```python
y4 = np.sin(x)
print(y4)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)

ax.plot(x, y1)

```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x,y1)
ax1.set_xlim([-5, 5])

ax2.plot(x,y2)

ax3.plot(x,y3)

ax4.plot(x,y4)
```
